### Data Visualization

```python
# data visualization using pairplot
import seaborn as sns
sns.pairplot(sample,hue='Region')
plt.show()
```

```python
#remove unnecessary coloumn such as postal code
sample = sample.drop(['Postal Code'],axis=1)
```

```python
# find pairwise correlation between coloumns and visualize using heatmap
sample.corr(numeric_only=True)
plt.figure(figsize=(12,6))
sns.heatmap(sample.corr(numeric_only=True), annot=True)
plt.show()
```

### Univariate Analysis

```python
sample['Ship Mode'].value_counts()
```

```python
sample['Ship Mode'].value_counts().plot(kind = 'pie', figsize=[10,10], autopct='%1.1f%%')
plt.title('Shipping mode wise order distribution')
```

Standard shipping is highly preferred while same day shipping is least preferred

```python
sample['Segment'].value_counts()
```

```python
sample['Segment'].value_counts().plot(kind= 'pie', figsize=[10,10],autopct='%1.1f%%')
plt.title('Segmentwise distribution')
```

we can see here that 52% people belong to consumer class whereas 17% to 30% belong to Home office & corporate

```python
sample['Country'].value_counts()
```

we can see that all the orders are under United States

```python
sample['Category'].value_counts()
```

```python
sample['Category'].value_counts().plot(kind='pie', figsize= [10,10], autopct='%1.1f%%')
plt.title('Categorywise distribution')
```

we can see that about 60% category belongs to office supplies whereas approxmately 20% category belongs to furniture and technological supplies 

```python
sample['Sub-Category'].value_counts()
```

```python
sample['Sub-Category'].value_counts().plot(kind='pie', figsize=[10,10], autopct='%1.1f%%')
plt.title('Sub-Category wise distribution')
```

Binders followed by papers are the most popular sub-category

```python
sample['City'].value_counts().head(20)
```

```python
sample['City'].value_counts().head(20).plot(kind='barh')
plt.title('Top 20 cities having most orders placed')
```

Maximum number of orders are placed in New York City

```python
sample['State'].value_counts().head()
```

```python
sample['State'].value_counts().sort_values(ascending= True).plot(kind='bar',figsize= [20,6])
plt.title('State wise order distribution')
```

Maximum orders are placed in California

```python
sample['Region'].value_counts()
```

```python
sample['Region'].value_counts().plot(kind='pie', figsize= [10,10], autopct= '%1.1f%%')
plt.title('Regional order distribution')
```

Western region has the maximum sales of 32%

### Bivariate Analysis

```python
categorical_sum= sample.groupby('Category').sum(numeric_only= True)
print(categorical_sum)
```

```python
categorical_sum[['Sales', 'Profit']].plot.bar()
```

Technological supplies have highest sales and profit. furnitures have moderate sales but least profit. Office supplies have least sales but moderate profit

```python
subcategorical_sum= sample.groupby('Sub-Category').sum(numeric_only= True)
subcategorical_sum
```

```python
subcategorical_sum[['Sales','Profit']].plot(kind= 'bar', figsize= [20,8])
plt.ylabel('Total Sales/Profit', fontsize= 15)
```

here tables have the lowest profit and copiers have the highest profit. also fasteners have the lowest sales whereas phones have the highest sales

```python
sample.groupby(by='State')['Profit'].sum().sort_values(ascending= True).plot(kind='bar',figsize= [20,8])
plt.ylabel('Profit', fontsize=15)
```

here we can see that maximum profit is in California and maximum loss is in Texas

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Discount', y='Profit', data=sample)
plt.title('Discount and profit')
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x='Discount', y='Sales', data=sample)
plt.title('Discount versus Sales')
```

It shows that if we give more discounts on products sales will increase but profits will fall
